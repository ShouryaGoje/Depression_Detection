```python
import pandas as pd
import numpy as np
import keras
from nrclex import NRCLex
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
# importing data
df = pd.read_csv('Analized.csv')
  
# head of the data
df.head()
  
X= df[['fear','anger','trust','sadness','disgust','negative']]
y=df['class_num']
  
# using the train test split function
x_train, x_test, y_train, y_test = train_test_split(X,y ,random_state=104,test_size=0.25, shuffle=True)

print(x_train, x_test, y_train, y_test)
```

```python
# Define the model
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=6))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# Train the model
model.fit(x_train, y_train, epochs=100, batch_size=64)
```

```python
# Evaluate the model
loss, accuracy = model.evaluate(x_test, y_test, batch_size=64)
print('Test loss:', loss)
print('Test accuracy:', accuracy)
```

```python
new_data = 'I feel like killing myself today, bye'

analysis = NRCLex(new_data).affect_frequencies
print(analysis)
new_df = pd.DataFrame(analysis, index=['stud1'])
new_df.head()
#Use the model to make predictions on the new data
predictions = model.predict(new_df[['fear','anger','trust','sadness','disgust','negative']])

# The predictions will be an array of probabilities for the positwive class
# You can convert them to binary class labels by thresholding them at 0.5
class_labels = (predictions > 0.5).astype(int)
print(class_labels)
```
