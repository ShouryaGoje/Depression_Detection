```python
# Limitation: NRCLex can't handle negations
# Limitation: NRCLex can't recognize words it doesn't know
from nrclex import NRCLex
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Depression_Detection.csv', encoding= 'unicode_escape')
df = df[['id','text','class']]
df.head()
print(df.dtypes)
```

```python
df['emotions'] = df['text'].apply(lambda x: NRCLex(str(x)).affect_frequencies)
df.head()

```

```python
df = pd.concat([df.drop(['emotions'], axis = 1), df['emotions'].apply(pd.Series)], axis = 1)
df.head()
```

```python
def binary(x):
    if x =='Depressed':
        return 1
    return 0
df['class_num'] = df['class'].apply(lambda x: binary(x))
df.head()
```

```python
df.to_csv('Analized.csv', index=False)
```
